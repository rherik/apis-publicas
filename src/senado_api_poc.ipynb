## Coletar Dados de Filiação dos Parlamentares

### Subtask:
Iterar sobre a lista de 'CodigoParlamentar' obtida. Para cada código, fazer uma requisição GET à API 'https://legis.senado.leg.br/dadosabertos/senador/{codigo}/filiacoes' para coletar os dados de filiação. Agrupar e estruturar esses dados, incluindo o nome do parlamentar e suas filiações, tratando possíveis erros de requisição.


**Reasoning**:
To efficiently retrieve the full name of each parliamentarian when processing their affiliation data, I will first create a mapping from `CodigoParlamentar` to `NomeCompletoParlamentar` from the already loaded `parlamentares_data`.



```python
parliamentarian_names = {}
for p in parlamentares_data:
    code = p['IdentificacaoParlamentar']['CodigoParlamentar']
    full_name = p['IdentificacaoParlamentar']['NomeCompletoParlamentar']
    parliamentarian_names[code] = full_name

print(f"First 5 parliamentarian name mappings: {list(parliamentarian_names.items())[:5]}")
```

**Reasoning**:
Now that the mapping of parliamentarian codes to full names is ready, the next step is to iterate through each parliamentarian code, make an API call to retrieve their affiliation data, extract the relevant fields, and store them in a structured list called `all_filiacoes_data`.



```python
all_filiacoes_data = []

for codigo_parlamentar in parliamentarian_codes:
    url_filiacoes = f'https://legis.senado.leg.br/dadosabertos/senador/{codigo_parlamentar}/filiacoes.json'

    try:
        response_filiacoes = requests.get(url_filiacoes)
        response_filiacoes.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
        filiacoes_json = response_filiacoes.json()

        # Get the full name of the parliamentarian using the mapping created earlier
        nome_completo_parlamentar = parliamentarian_names.get(codigo_parlamentar, 'Nome Desconhecido')

        filiacoes = filiacoes_json.get('DetalheParlamentar', {}).get('Parlamentar', {}).get('Filiacoes', {}).get('Filiacao', [])

        # Ensure filiacoes is always a list, even if there's only one affiliation
        if isinstance(filiacoes, dict):
            filiacoes = [filiacoes]

        if not filiacoes:
            # Handle cases where no affiliation data is found
            all_filiacoes_data.append({
                'Nome Parlamentar': nome_completo_parlamentar,
                'Codigo Parlamentar': codigo_parlamentar,
                'Sigla Partido': '',
                'Data Filiacao': '',
                'Data Desfiliacao': ''
            })
            continue

        for filiacao in filiacoes:
            filiacao_data = {
                'Nome Parlamentar': nome_completo_parlamentar,
                'Codigo Parlamentar': codigo_parlamentar,
                'Sigla Partido': filiacao.get('SiglaPartido', ''),
                'Data Filiacao': filiacao.get('DataFiliacao', ''),
                'Data Desfiliacao': filiacao.get('DataDesfiliacao', '')
            }
            all_filiacoes_data.append(filiacao_data)

    except requests.exceptions.RequestException as e:
        print(f"Erro na requisição para {codigo_parlamentar}: {e}")
        all_filiacoes_data.append({
            'Nome Parlamentar': parliamentarian_names.get(codigo_parlamentar, 'Nome Desconhecido'),
            'Codigo Parlamentar': codigo_parlamentar,
            'Sigla Partido': 'Erro',
            'Data Filiacao': 'Erro',
            'Data Desfiliacao': 'Erro'
        })

print(f"Total affiliation records collected: {len(all_filiacoes_data)}")
if all_filiacoes_data:
    print(f"First 3 affiliation records: {all_filiacoes_data[:3]}")
```

```python
import requests
import pandas as pd

url_base = 'https://dadosabertos.camara.leg.br/api/v3'
session = requests.Session()
```

```python

def deputados_para_dataframe(deputados) -> pd.DataFrame:
  """Converte lista de deputados para DataFrame"""
  dados_processados = []
  for deputado in deputados:
      try:
          dados_processados.append({
              'id': deputado.get('id'),
              'uri': deputado.get('uri'),
              'nome': deputado.get('nome'),
              'siglaPartido': deputado.get('siglaPartido'),
              'uriPartido': deputado.get('uriPartido'),
              'siglaUf': deputado.get('siglaUf'),
              'idLegislatura': deputado.get('idLegislatura'),
              'urlFoto': deputado.get('urlFoto'),
              'email': deputado.get('email')
          })
      except Exception as e:
          print(f"Erro ao processar deputado: {e}")
  df = pd.DataFrame(dados_processados)
  return df

# session.mount('https://', TLSv1_2Adapter())
# session.headers.update({'Accept': 'application/json',
# 'User-Agent': 'Python-API-Senado/1.0'})

"""
Busca lista de todos os deputados em exercício
Se não for passado um parâmetro de tempo, como idLegislatura ou dataInicio, a lista enumerará somente os deputados em exercício no momento da requisição.
"""
endpoint = f"/deputados"
url = f"{url_base}{endpoint}"
params = {'dataInicio': '2024-01-01', 'dataFim': '2024-12-31'}

try:
    if params is not None:
        response = session.get(url, params=params, timeout=25)
    else:
        response = session.get(url, timeout=25)
    response.raise_for_status()
    #return response.json()
except requests.exceptions.Timeout:
    print(f"Tempo limite exedido para {url}")
except requests.exceptions.RequestException as e:
    print(f"Erro na requisição para {url}: \n {e}")


deputados = response.json()
df_deputados = deputados_para_dataframe(deputados['dados'])
print(f"Encontrados {len(df_deputados)} deputados.")
for index, deputado in df_deputados.iterrows():
    print(f"{index+1} - {deputado['nome']} - {deputado['id']} - {deputado['siglaPartido']} - {deputado['idLegislatura']} - {deputado['siglaUf']}")
```

## **Busca despesas**

Dá acesso aos registros de pagamentos e reembolsos feitos pela Câmara em prol do deputado identificado por {id}, a título da Cota para Exercício da Atividade Parlamentar, a chamada "cota parlamentar".

A lista pode ser filtrada por mês, ano, legislatura, CNPJ ou CPF de um fornecedor.
Se não forem passados os parâmetros de tempo, o serviço retorna os dados dos seis meses anteriores à requisição.

```python
def despesas_para_dataframe(despesas) -> pd.DataFrame:
  """Converte lista de despesas retornadas para DataFrame"""
  dados_processados = []
  for despesa in despesas:
      try:
          dados_processados.append({
              'ano': despesa.get('ano'),
              'mes': despesa.get('mes'),
              'tipoDespesa': despesa.get('tipoDespesa'),
              'codDocumento': despesa.get('codDocumento'),
              'tipoDocumento': despesa.get('tipoDocumento'),
              'codTipoDocumento': despesa.get('codTipoDocumento'),
              'dataDocumento': despesa.get('dataDocumento'),
              'numDocumento': despesa.get('numDocumento'),
              'valorDocumento': despesa.get('valorDocumento'),
              'urlDocumento': despesa.get('urlDocumento'),
              'nomeFornecedor': despesa.get('nomeFornecedor'),
              'cnpjCpfFornecedor': despesa.get('cnpjCpfFornecedor'),
              'valorLiquido': despesa.get('valorLiquido'),
              'valorGlosa': despesa.get('valorGlosa'),
              'numRessarcimento': despesa.get('numRessarcimento'),
              'codLote': despesa.get('codLote')
          })

      except Exception as e:
          print(f"Erro {e} ao processar despesas.")
      df = pd.DataFrame(dados_processados)
      return df

def busca_despesas(id):
    endpoint = f"/deputados/{id}/despesas"
    url = f"{url_base}{endpoint}"
    try:
        response = session.get(url, timeout=25)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.Timeout:
        print(f"Tempo limite exedido para {url}")
    except requests.exceptions.RequestException as e:
        print(f"Erro na requisição para {url}: \n {e}")

for id_dep, row in df_deputados.iterrows():
  if nome.lower() in row["nome"].lower():
      print(f"{row["nome"]} de ID {id_dep} encontrado(a) para exibir as despesas.\n")

      despesas = busca_despesas(id=row["id"])
      df_despesas = despesas_para_dataframe(despesas['dados'])

      print(df_despesas)
```
